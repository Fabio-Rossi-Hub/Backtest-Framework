# Backtesting of a pair trading strategy

```python
from backtesting.utils import FuturesPairDataset
from backtesting.backtest import Backtest
```

### Data Import

```python
dataset = FuturesPairDataset(path_a='example_data/futuresA.csv', path_b='example_data/futuresB.csv')

dataset.df.tail()
```

### Backtest

```python
# Set initial values for parameters
L = 100  # Long position size
S = 100  # Short position size
X = 2    # Standard deviation drop threshold
Y = 2    # Standard deviation rise threshold
C = 0.1  # Transaction cost
N = 5    # Number of days for standard deviation calculation

# Create a Backtest instance
backtest = Backtest(
    dataset=dataset,  # Dataset containing historical price data
    strategy='future_pairs_strategy',  # Trading strategy to be backtested
    n_days=N,  # Number of days for standard deviation calculation
    asset_a='A Price',  # Name of the first asset
    asset_b='B Price',  # Name of the second asset
    std_rise=Y,  # Standard deviation rise threshold
    std_drop=X,  # Standard deviation drop threshold
    C=C,  # Transaction cost
    long_size=L,  # Long position size
    short_size=S  # Short position size
)
```

```python
backtest.summary()
```

```python
backtest.plot_pnl()
```

```python
backtest.plot_drawdown()
```
